```python
import argparse
import glob
import os
from pathlib import Path

import nibabel as nib
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from dng.dataset.hcp import HCPDataset
from dng.models.models3D import ResNetAutoencoder3d
from torch.utils.data import DataLoader, random_split
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)

model = ResNetAutoencoder3d(num_channels=64, nonlinearity=F.leaky_relu).to(device)
ckpt = "hcp3d_autoencoder_checkpoint_epoch079.pth"
checkpoint = torch.load(ckpt)
model.load_state_dict(checkpoint["model"])
model.eval()

encodings_path = Path(
    "/home/sz9jt/manifold/sz9jt/generative_brain/diffusion/diffusion_tomhcp3dlatent_4ch/sample-ckpt750"
)
out_path = encodings_path.parent / "sample_decoded_ckpt750"
out_path.mkdir(exist_ok=True)

filelist = glob.glob(str(encodings_path / "*unscaled.npy"))
filelist = [str(x).split("/")[-1] for x in filelist]
print(len(filelist))
print(filelist)

for file in filelist:
    if os.path.exists(out_path / file.replace("_unscaled.npy", ".nii.gz")):
        print(file)
        continue
    data = np.load(encodings_path / file)
    z = torch.from_numpy(data).float().to(device)
    if len(z.shape) != 5:
        z = z.unsqueeze(0)
    with torch.no_grad():
        x = model.decoder(z)
    print(x.shape)
    img = nib.Nifti1Image(np.squeeze(x.data.cpu().numpy()), affine=np.eye(4))
    nib.save(img, out_path / file.replace("_unscaled.npy", ".nii.gz"))
    del x
    del z
```
