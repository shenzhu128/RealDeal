```python
from pathlib import Path

import sys

sys.path.append("..")

import pandas as pd
import torch
from generative.metrics import MultiScaleSSIMMetric
from generative.networks.nets import AutoencoderKL
from monai.config import print_config
from monai.utils import set_determinism
from omegaconf import OmegaConf
from tqdm import tqdm
import matplotlib.pyplot as plt
import nibabel as nib
import numpy as np
import os

from models.autoencoderkl import AutoencoderKLDownsampleControl

from datasets import combined_brain_1mm, combined_brain_17mm
from torch.utils.data import DataLoader
```

```python
def get_stage1_model(path, index, device="cuda:0"):
    stage1_path = Path(path)
    stage1_ckpt = stage1_path / f"checkpoint{index}.pth"
    stage1_config = stage1_path / "config.yaml"
    
    stage1_config = OmegaConf.load(stage1_config)

    print("Creating model...")
    device = torch.device(device)
    stage1 = AutoencoderKLDownsampleControl(**stage1_config["stage1"]["params"])
    stage1 = torch.nn.DataParallel(stage1)

    stage1.load_state_dict(torch.load(stage1_ckpt)["state_dict"])
    stage1 = stage1.to(device)
    stage1.eval()

    print("Stage1 Model created")
    
    return stage1, device


def decode_brain_latent_codes(diffusion_path, stage1, device, scale=0.2, source_dir="sample-npy", output_dir="sample-nifti"):
    # Diffusion for brain aekl 4x 4ch
    # Diffusion output 48, 56, 48
    # Actually encoding size 44, 52, 46
    
    # diffusion for brain aekl 2x 3ch
    # diffusion output 88, 104, 96
    # actually encoding size 88, 104, 92
    
    # diffusion for brain aekl 8x 3ch
    # diffusion output 24 32 24
    # actually encoding size 22, 26, 23 (right pad)
    diffusion_path = Path(diffusion_path)
    latent_samples = os.listdir(diffusion_path / source_dir)
    latent_samples = [item for item in latent_samples if item.endswith(".npy")]
    latent_samples.sort()
    print(f"Number of latent samples {len(latent_samples)}")
    
    for latent_sp in latent_samples:
        data = np.load(diffusion_path / f"{source_dir}/{latent_sp}")
        data = data[:, :, 1:-1, 3:-3, :-1]
        data = data / scale
        print(f"Latent sample {latent_sp}")
        print(data.shape, data.min(), data.max())

        input = torch.from_numpy(data).float().to(device)
        with torch.no_grad():
            output = stage1.module.decode(input)
        output_numpy = output.cpu().numpy().squeeze()
        output_name = latent_sp.replace(".npy", "")
        if len(output_numpy.shape) == 3:
            output_nifti = nib.Nifti1Image(output_numpy, np.eye(4))
            nib.save(output_nifti, diffusion_path/f"{output_dir}/{output_name}.nii.gz")
        else:
            for i in range(len(output_numpy)):
                output_nifti = nib.Nifti1Image(output_numpy[i], np.eye(4))
                nib.save(output_nifti, diffusion_path/f"{output_dir}/{output_name}_{i}.nii.gz")
```

```python
diffusion_path = Path("/home/sz9jt/data/generative_brain/diffusion/diffusion_smaller_ldmp_3ch_8x_brain_corrected_1_attention")
diffusion_config_path = diffusion_path / "config.yaml"
diffusion_config = OmegaConf.load(diffusion_config_path)

stage1_path = Path(diffusion_config.dataset.params.data_paths) / ".."
ckpt = diffusion_config.dataset.params.data_paths.split("/")[-1]
ckpt = ckpt.split("_")[1].replace("ckpt", "")
ckpt = int(ckpt)
scale = diffusion_config.dataset.params.fixed_scale

stage1, device = get_stage1_model(stage1_path, ckpt)
decode_brain_latent_codes(diffusion_path, stage1, device, scale=scale, source_dir="sample-ckpt100", output_dir="sample-ckpt100-decoded")
# decode_brain_latent_codes(diffusion_path, stage1, device, scale=scale)
```
