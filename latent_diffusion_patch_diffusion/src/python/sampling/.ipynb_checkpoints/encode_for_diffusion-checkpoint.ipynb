```python
import sys

sys.path.append("..")

import torch
import torch.nn as nn
from datasets import combined_brain_1mm, combined_brain_17mm
from generative.networks.nets import AutoencoderKL
from models.autoencoderkl import AutoencoderKLDownsampleControl
from omegaconf import OmegaConf
import numpy as np
from tqdm import tqdm
import os
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from pathlib import Path
import nibabel as nib
```

```python
def get_brain_datasets(root_dir, corrected):
    train_ds = combined_brain_1mm.CombinedBrainDataset(root_dir=root_dir, type="train", return_name=True, corrected=corrected)
    val_ds = combined_brain_1mm.CombinedBrainDataset(root_dir=root_dir, type="val", return_name=True, corrected=corrected)
    test_ds = combined_brain_1mm.CombinedBrainDataset(root_dir=root_dir, type="test", return_name=True, corrected=corrected)
    return train_ds, val_ds, test_ds
    
def get_stage1_model(config_path, weight_path):
    config = OmegaConf.load(config_path)
    print("Creating model")
    model = AutoencoderKLDownsampleControl(**config["stage1"]["params"])
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    data = torch.load(weight_path)
    model = nn.DataParallel(model)
    model.to(device)
    model.load_state_dict(data["state_dict"])
    print("Model created")
    
    return model, device

def encode_dataset(ds, output_path, device, model):
    for i, _ in enumerate(tqdm(ds)):
        input = ds[i]["data"].unsqueeze(0).to(device)
        names = ds[i]["name"]
        with torch.no_grad():
            z_mu, z_sigma = model.module.encode(input)
        output = np.concatenate([z_mu.cpu().numpy().squeeze(), z_sigma.cpu().numpy().squeeze()])
        if i == 0:
            print(output.shape)
        np.save(os.path.join(output_path, names+".npy"), output)
        
        
def visualize_latent_code(output_path, ch=3):
    encodings = os.listdir(output_path)
    encodings = [item for item in encodings if item.endswith(".npy")]
    for i in range(10):
        latent = np.load(output_path / encodings[i])
        nrow = latent.shape[0] // ch
        ncol = ch
        fig = plt.figure()
        # row 1
        for j in range(ncol):
            ax = fig.add_subplot(nrow + 1, ncol, j + 1)
            ax.imshow(latent[j, 20], cmap="gray")
            ax.set_title(f"z_mu ch{j}")
            ax.axis("off")
        # row 2
        for j in range(ncol):
            ax = fig.add_subplot(nrow + 1, ncol, ncol * 1 + j + 1)
            ax.imshow(latent[j + ncol, 20], cmap="gray")
            ax.set_title(f"z_sigma ch{j}")
            ax.axis("off") 
        sample = np.random.randn(*latent[:ch].shape)
        sample = latent[:ch] + sample * latent[ch:]
        for j in range(ncol):
            ax = fig.add_subplot(nrow + 1, ncol, nrow * ncol + j + 1)
            ax.imshow(sample[j, 20], cmap="gray")
            ax.set_title(f"sampled ch{j}")
            ax.axis("off")
        out = "../latent_code_visualization/"+encodings[i].replace(".npy", ".png")
        plt.savefig(output_path / out)
        plt.close()
        
        plt.hist(sample.flatten(), bins=1000, density=True)
        x = np.linspace(-4, 4, 100)
        y = np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)
        plt.plot(x, y)
        out = "../latent_code_visualization/distribution.png"
        plt.savefig(output_path / out)
        plt.close()
        
        out = "../latent_code_visualization/"+encodings[i]
        np.save(output_path / out, sample)
        break
```

```python
path = Path("/home/sz9jt/data/generative_brain/stage1/vanilla_aekl_corrected_v1_4x_3ch")
ckpt = 128

config_path = path / "config.yaml"
weight_path = path / f"checkpoint{ckpt}.pth"
output_path = path / f"encodings_ckpt{ckpt}_best_val_loss"

# root_dir = "/home/sz9jt/magpie/outputs"
root_dir = "/home/sz9jt/data/t1w_processed/outputs_corrected"

train_ds, val_ds, test_ds = get_brain_datasets(root_dir, corrected=True)
model, device = get_stage1_model(config_path, weight_path)

encode_dataset(train_ds, output_path, device, model)
encode_dataset(val_ds, output_path, device, model)
```

```python
import matplotlib.pyplot as plt

data = np.load("/home/sz9jt/data/generative_brain/stage1/aekl_v2_4x_4ch/encodings_ckpt172_best_val_loss/abide1.sub-0050039.ses-Pitt.EMPTY.npy")
z_mu = data[:4]
z_sigma = data[4:]

eps = np.random.randn(*z_mu.shape)
z_mu = z_mu + eps * z_sigma

fig = plt.figure()
for i in range(len(z_mu)):
    ax = fig.add_subplot(1, 4, i + 1)
    ax.imshow(z_mu[i, 24], cmap="gray")
    ax.axis("off")
plt.show()
```

```python
print(z_mu.min(), z_mu.max())
print(z_sigma.min(), z_sigma.max())
mean_z_mu = z_mu.mean()
mean_z_sigma = z_sigma.mean()
print(mean_z_mu, mean_z_sigma)

tmp_mu = torch.ones_like(z_mu) * mean_z_mu
tmp_sigma = torch.ones_like(z_sigma) * mean_z_sigma
print(tmp_mu.shape, tmp_sigma.shape)
z_vae = model.module.sampling(tmp_mu, tmp_sigma)
print(z_vae.shape)
```

```python
device = torch.device("cuda:1")
model.to(device)

with torch.no_grad():
    z_vae = model.module.sampling(z_mu, z_sigma * 3)
    output = model.module.decode(z_vae.to(device))
    print(output.shape)
```
