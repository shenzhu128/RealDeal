```python
import os
import numpy as np
import matplotlib.pyplot as plt
import math
import nibabel as nib
from pathlib import Path
import subprocess
```

```python
def get_range(output_path, num=1000, ch=4, sample=False):
    encodings = os.listdir(output_path)
    print(len(encodings))

    idx = np.random.randint(0, len(encodings), num)
    min_values = []
    max_values = []
    for i in idx:
        # data = np.load(os.path.join(output_path, encodings[i]))[:ch]
        data = np.load(os.path.join(output_path, encodings[i]))
        if sample:
            img_mu = data[:ch]
            img_sigma = data[ch:]
            eps = np.random.randn(*img_mu.shape)
            data = img_mu + eps * img_sigma
        else:
            data = data[:ch]
        min_values.append(data.min())
        max_values.append(data.max())
    print(data.shape)
    plt.hist(min_values, bins=100)
    plt.hist(max_values, bins=100)
    plt.savefig(os.path.join(output_path, "..", "encoding_range.png"))


def get_range_8mm_nifty(output_path, num=1000, sample=False):
    encodings = os.listdir(output_path)
    encodings = [x for x in encodings if x.endswith(".nii.gz")]
    print(len(encodings))

    idx = np.random.randint(0, len(encodings), num)
    min_values = []
    max_values = []
    for i in idx:
        data = nib.load(os.path.join(output_path, encodings[i])).get_fdata()
        if sample:
            img_sigma = nib.load(os.path.join(output_path, "sigma", encodings[i])).get_fdata()
            eps = np.random.randn(*img_sigma.shape)
            data = data + eps * img_sigma
        
        min_values.append(data.min())
        max_values.append(data.max())
    print(data.shape)
    plt.hist(min_values, bins=100)
    plt.hist(max_values, bins=100)
    plt.savefig(os.path.join(output_path, "..", "encoding_range.png"))
    return min_values, max_values
    

def get_nifti_range(output_path, num=1000):
    output_path = Path(output_path)
    encodings = os.listdir(output_path)
    print(len(encodings))
    
    idx = np.random.randint(0, len(encodings), num)
    min_values = []
    max_values = []
    for i in idx:
        data = nib.load(output_path / encodings[i]).get_fdata()
        min_values.append(data.min())
        max_values.append(data.max())
    print(data.shape)
    plt.hist(min_values, bins=100)
    plt.hist(max_values, bins=100)
    plt.savefig(os.path.join(output_path, "..", "encoding_range.png"))
    

def get_range_per_channel_8mm_nifty(output_path, num=1000, sample=False):
    encodings = os.listdir(output_path)
    encodings = [x for x in encodings if x.endswith(".nii.gz")]
    print(len(encodings))

    if num >= len(encodings):
        idx = np.arange(0, len(encodings))
    else:
        idx = np.random.randint(0, len(encodings), num)
    print("len idx", len(idx))
    min_values = []
    max_values = []

    # Load a sample file to determine the number of channels
    tmp = nib.load(os.path.join(output_path, encodings[0])).get_fdata()
    num_ch = tmp.shape[-1]

    print("Encoding channels: ", num_ch)

    # Initialize lists to store min and max values for each channel
    chwise_min_values = [[] for _ in range(num_ch)]
    chwise_max_values = [[] for _ in range(num_ch)]
    
    for i in idx:
        data = nib.load(os.path.join(output_path, encodings[i])).get_fdata()
        if sample:
            img_sigma = nib.load(os.path.join(output_path, "sigma", encodings[i])).get_fdata()
            eps = np.random.randn(*img_sigma.shape)
            data = data + eps * img_sigma
        min_values.append(data.min())
        max_values.append(data.max())

        # how to handle uncertain multiple channels here
        for ch in range(num_ch):
            chwise_min_values[ch].append(data[..., ch].min())
            chwise_max_values[ch].append(data[..., ch].max())

    print(data.shape)

    tmp = np.array(chwise_min_values)
    print(tmp.shape)
    tmp = np.array(chwise_max_values)
    print(tmp.shape)
    tmp = np.array(min_values)
    print(tmp.shape)

    plt.hist(min_values, bins=100)
    plt.hist(max_values, bins=100)
    plt.savefig(os.path.join(output_path, "..", "encoding_range.png"))
    plt.close()

    for i in range(num_ch):
        plt.hist(chwise_min_values[i], bins=100)
        plt.hist(chwise_max_values[i], bins=100)
        plt.savefig(os.path.join(output_path, "..", f"c{i+1}_encoding_range.png"))
        plt.close()
```
