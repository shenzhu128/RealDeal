```python
import sys
from pathlib import Path

sys.path.append("..")

import os

import matplotlib.pyplot as plt
import nibabel as nib
import numpy as np
import pandas as pd
import torch

import datasets
from datasets import brain_5_datasets_1mm, combined_brain_1mm, combined_brain_17mm, hcp_3d
from generative.metrics import MultiScaleSSIMMetric
from generative.networks.nets import AutoencoderKL
from models.autoencoderkl import AutoencoderKLDownsampleControl
from monai.config import print_config
from monai.utils import set_determinism
from omegaconf import OmegaConf
from torch.utils.data import DataLoader
from tqdm import tqdm
```

```python
def get_config(config_path):
    return OmegaConf.load(config_path)


def get_stage1_model(config_path, stage1_ckpt_path):
    config = OmegaConf.load(config_path)
    stage1_path = stage1_ckpt_path

    set_determinism(seed=config.args.seed)

    print("Creating model...")
    device = torch.device("cuda")
    stage1 = AutoencoderKLDownsampleControl(**config["stage1"]["params"])
    stage1 = torch.nn.DataParallel(stage1)

    stage1.load_state_dict(torch.load(stage1_path)["state_dict"])
    stage1 = stage1.to(device)
    stage1.eval()
    print("Model created")

    return stage1


def reconstruct_images(
    result_folder,
    dataset,
    output_size=(192, 224, 192),
):
    config = OmegaConf.load(result_folder / "config.yaml")
    best_ckpt = config.best_ckpt.index
    stage1_path = result_folder / f"checkpoint{best_ckpt}.pth"
    outpath = result_folder / f"decoded_imgs_ckpt{best_ckpt}"
    # if outpath does not exist, create it
    outpath.mkdir(exist_ok=True)

    set_determinism(seed=config.args.seed)

    print(f"Using checkpoitn {best_ckpt}")
    print("Creating model...")
    device = torch.device("cuda")
    stage1 = AutoencoderKLDownsampleControl(**config["stage1"]["params"])
    stage1 = torch.nn.DataParallel(stage1)

    stage1.load_state_dict(torch.load(stage1_path)["state_dict"])
    stage1 = stage1.to(device)
    stage1.eval()
    print("Model created")

    loader = DataLoader(
        dataset,
        batch_size=4,
        shuffle=False,
        num_workers=2,
        drop_last=False,
        pin_memory=False,
        persistent_workers=True,
    )

    for batch in tqdm(loader):
        x = batch["data"].to(device)
        names = batch["name"]

        with torch.no_grad():
            x_recon, _, _ = stage1.module(x)

        outimg = x_recon.cpu().numpy().squeeze()
        assert outimg.shape[1:] == output_size
        for i, single in enumerate(outimg):
            single = nib.Nifti1Image(single, np.eye(4))
            nib.save(single, os.path.join(outpath, names[i] + ".nii.gz"))
```
