### For regular LDM models

```python
def get_diffusion_model(diffusion_root_path, ckpt):
    diffusion_root_path = Path(diffusion_root_path)
    diffusion_config_path = diffusion_root_path / "config.yaml"
    diffusion_ckpt_path = diffusion_root_path / f"checkpoint{ckpt}.pth"

    config = OmegaConf.load(diffusion_config_path)
    diffusion_model = DiffusionModelUNet(**config["ldm"].get("params", dict()))
    scheduler = DDPMScheduler(**config["ldm"].get("scheduler", dict()))
    diffusion_ckpt = torch.load(diffusion_ckpt_path)

    diffusion_model = torch.nn.DataParallel(diffusion_model)
    diffusion_model.load_state_dict(diffusion_ckpt["diffusion"])

    output_shape = config.dataset.params.output_shape
    # ch = config.dataset.params.ch
    ch = 3
    output_shape = (ch,) + tuple(output_shape)
    scale_factor = config.dataset.params.fixed_scale

    return diffusion_model, scheduler, output_shape, scale_factor


@torch.no_grad()
def sampling_procedure(
    outpath, diffusion_model, scheduler, device, output_shape, scale_factor, idx, num_of_samples, seed
):
    latent = []
    torch.manual_seed(seed)
    tmp = torch.randn(output_shape)
    latent.append(tmp)
    torch.manual_seed(seed + 100)
    tmp = torch.randn(output_shape)
    latent.append(tmp)
    latent = torch.stack(latent)
    latent = latent.to(device)

    prompt_embeds = None
    for t in tqdm(scheduler.timesteps, ncols=70):
        noise_pred = diffusion_model(x=latent, timesteps=torch.asarray((t,)).to(device), context=prompt_embeds)
        latent, _ = scheduler.step(noise_pred, t, latent)

    #     nrow = len(latent)
    #     ncol = len(latent[0])
    #     x_slice = 24

    #     fig = plt.figure()
    #     for j in range(nrow * ncol):
    #         ax = fig.add_subplot(nrow, ncol, j + 1)
    #         ax.axis("off")
    #         ax.imshow(latent[j // ncol, j % ncol, x_slice].cpu().numpy(), cmap="gray")
    #     plt.savefig(str(outpath / f"sample_epoch{epoch}_{i}.png"), bbox_inches="tight")
    #     plt.close()

    # np.save(str(outpath / f"sample_epoch{epoch}_fixscale{scale_factor}_{idx}.npy"), latent.cpu().numpy())
    return latent
```

```python
diffusion_root_path = Path("/home/sz9jt/data/generative_brain/diffusion/diffusion_3ch_relu_8mm")
ckpt = 1000

diffusion_model, scheduler, output_shape, scale_factor = get_diffusion_model(diffusion_root_path, ckpt)
print(f"root path: {diffusion_root_path} \n output shape {output_shape} \n scale factor {scale_factor}")
outpath = diffusion_root_path / f"diffusion_process"
os.makedirs(outpath, exist_ok=True)

device = torch.device("cuda:0")

with torch.no_grad():
    latent = torch.randn((1,) + output_shape)
    latent = latent.to(device)

    prompt_embeds = None
    for t in tqdm(scheduler.timesteps, ncols=70):
        noise_pred = diffusion_model(x=latent, timesteps=torch.asarray((t,)).to(device), context=prompt_embeds)
        latent, _ = scheduler.step(noise_pred, t, latent)
        if t % 100 == 0:
            data = latent.cpu().numpy()
            plt.axis("off")
            plt.imshow(np.rot90(data[0, 1, :, :, 30]), cmap="gray")
            plt.savefig(outpath / f"timestep{t}.png", bbox_inches="tight")
        if t < 100 and t % 10 == 0:
            data = latent.cpu().numpy()
            plt.axis("off")
            plt.imshow(np.rot90(data[0, 1, :, :, 30]), cmap="gray")
            plt.savefig(outpath / f"timestep{t}.png", bbox_inches="tight")
```

### For conditional diffusion models

```python
def get_cond_diffusion_model(diffusion_root_path, ckpt):
    diffusion_root_path = Path(diffusion_root_path)
    diffusion_config_path = diffusion_root_path / "config.yaml"
    diffusion_ckpt_path = diffusion_root_path / ckpt
    print(f"Diffusion path {diffusion_ckpt_path}")

    config = OmegaConf.load(diffusion_config_path)
    diffusion_model = DiffusionModelUNet(**config["ldm"].get("params", dict()))
    scheduler = DDPMScheduler(**config["ldm"].get("scheduler", dict()))
    diffusion_ckpt = torch.load(diffusion_ckpt_path)

    diffusion_model = torch.nn.DataParallel(diffusion_model)
    # diffusion_model.load_state_dict(diffusion_ckpt["diffusion"])
    diffusion_model.load_state_dict(diffusion_ckpt["diffusion"])

    checkpoint_name = str(diffusion_ckpt_path).split("/")[-1]

    return diffusion_model, scheduler, checkpoint_name
```

```python
device = torch.device("cuda")
root_path = "/home/sz9jt/data/generative_brain/cond_diffusion/diffusion_hcp3d_cond"
diffusion_model, scheduler = get_cond_diffusion_model(root_path, 100)
diffusion_model.to(device)

config = OmegaConf.load(root_path + "/config.yaml")
ds_params = config.dataset.params
ds_params["type"] = "test"
test_ds = datasets.hcp_3d.HCP3DCondDataset(**ds_params)

data = test_ds[-2]
print(data["gt_image"].shape, data["cond_image"].shape)
cond = torch.tensor(data["cond_image"]).unsqueeze(0).to(device)
print("cond", cond.shape)

torch.manual_seed(config.args.seed)

with torch.no_grad():
    y = torch.randn(data["gt_image"].shape).unsqueeze(0).to(device)

    prompt_embeds = None
    for t in tqdm(scheduler.timesteps, ncols=70):
        tmp_input = torch.cat([y, cond], dim=1).float().to(device)
        noise_pred = diffusion_model(x=tmp_input, timesteps=torch.asarray((t,)).to(device), context=prompt_embeds)
        y, _ = scheduler.step(noise_pred, t, y)
        # if t % 200 == 0:
        #     tmp = y.cpu().numpy()
        #     plt.axis("off")
        #     plt.imshow(np.rot90(tmp[0, 0, :, :, 30]), cmap="gray")
        #     plt.show()

gt_img = data["gt_image"].squeeze()
cond_img = data["cond_image"].squeeze()
out_img = y.cpu().numpy().squeeze()

fig, axs = plt.subplots(8, 3, figsize=(6, 18))

for i in range(0, 64, 8):
    axs[i // 8, 0].imshow(np.rot90(gt_img[:, :, i]), cmap="gray")
    axs[i // 8, 0].set_title("gt")
    axs[i // 8, 0].axis("off")

    axs[i // 8, 1].imshow(np.rot90(cond_img[:, :, i]), cmap="gray")
    axs[i // 8, 1].set_title("cond")
    axs[i // 8, 1].axis("off")

    axs[i // 8, 2].imshow(np.rot90(out_img[:, :, i]), cmap="gray")
    axs[i // 8, 2].set_title("output")
    axs[i // 8, 2].axis("off")

plt.show()
```

```python
import os
from pathlib import Path

import numpy as np

# unscale toms latent encodings
cutoff_vals = [[0.2, 0.9], [0.1, 0.65], [0.1, 0.85], [0.2, 1.0]]


def scale_to_range(data, leftval, rightval):
    data = (data - leftval) / (rightval - leftval)
    data = data * 2 - 1
    return data


def unscale_to_orig(data, leftval, rightval):
    data = (data + 1) / 2  # back to 0-1
    data = (data * (rightval - leftval)) + leftval  # back t0 leftval to rightval
    return data


rootpath = Path("/home/sz9jt/data/generative_brain/diffusion/diffusion_tomhcp3dlatent_4ch/sample-npy")
items = os.listdir(rootpath)
items = [x for x in items if not x.endswith("unscaled.npy")]
print(items)

for file in items:
    data = np.load(rootpath / file).squeeze()
    print(data.min(), data.max())
    print(data.shape)
    output = np.zeros(data.shape)
    for i in range(len(output)):
        output[i] = unscale_to_orig(data[i], *cutoff_vals[i])
        print(output[i].min(), output[i].max())
    np.save(rootpath / file.replace(".npy", "_unscaled.npy"), output)
```
