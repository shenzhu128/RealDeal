```python
import os
import torch
import numpy as np
from pathlib import Path
```

```python
def get_ckpts_best_loss(path):
    ckpts = os.listdir(path)
    ckpts = [item for item in ckpts if item.startswith("checkpoint")]
    num_ckpts = len(ckpts)

    res = []
    for i in range(1, num_ckpts + 1):
        if torch.cuda.is_available():
            data = torch.load(os.path.join(path, f"checkpoint{i}.pth"))
        else:
            data = torch.load(os.path.join(path, f"checkpoint{i}.pth"), map_location=torch.device("cpu"))
        best_loss = data["best_loss"]
        res.append(best_loss)
        print(f"checkpoint{i}.pth", best_loss)

    return res, ckpts


def get_ckpts_best_loss_list(path):
    ckpts = os.listdir(path)
    ckpts = [item for item in ckpts if item.startswith("checkpoint")]
    ckpts.sort()
    res = []
    
    for ckpt_name in ckpts:
        if torch.cuda.is_available():
            data = torch.load(os.path.join(path, ckpt_name))
        else:
            data = torch.load(os.path.join(path, ckpt_name), map_location=torch.device("cpu"))
        best_loss = data["best_loss"]
        ckpt_num = ckpt_name.replace(".pth", "").replace("checkpoint", "")
        if len(ckpt_num) == 0:
            continue
        ckpt_num = int(ckpt_num)
        res.append([ckpt_num, best_loss])
        # print(ckpt_num, best_loss)

    return res
```

```python
path = Path("/home/sz9jt/data/generative_brain/stage1/vanilla_aekl_corrected_v1_4x_3ch")
image_path = path / "images"
all_images = os.listdir(image_path)
all_images = [item for item in all_images if item.endswith(".png")]

loss_res = []
for item in all_images:
    tmp = item.split("_")
    index = int(tmp[2])
    loss = tmp[-1].replace(".png", "")
    loss = float(loss[4:])
    loss_res.append([index, loss])
loss_res = sorted(loss_res, key=lambda x: x[1])
print(loss_res)
```
