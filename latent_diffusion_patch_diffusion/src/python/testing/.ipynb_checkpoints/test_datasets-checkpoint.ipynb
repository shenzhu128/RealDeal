```python
import sys

sys.path.append("../")
sys.path.append("../training")

from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import training.util
from datasets import (
    brain_5_datasets_1mm,
    combined_brain_1mm,
    combined_brain_17mm,
    inr_encodings,
    ldmp_encodings,
    lidc,
)
from omegaconf import OmegaConf
from torch.utils.data import DataLoader
```

```python
root_dir = "/home/sz9jt/data/t1w_processed/outputs"
ds = brain_5_datasets_1mm.CombinedBrainDataset(root_dir=root_dir, type="all")

print(ds[0]["data"].shape)
print(ds[0]["data"].min(), ds[0]["data"].max())

train_loader = DataLoader(
    ds,
    batch_size=2,
    shuffle=True,
    num_workers=2,
    drop_last=False,
    pin_memory=False,
    persistent_workers=True,
)
batch = next(iter(train_loader))
print(batch["data"].shape)
```
